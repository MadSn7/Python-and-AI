# Common probability distributions 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu, sigma = 0, 1

# Values
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

# Plot PDF
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('Density')

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, norm.cdf(x, mu, sigma))
plt.title('Normal Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import expon

# Parameters
lambda_ = 1

# Values
x = np.linspace(0, 5, 100)

# Plot PDF
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, expon.pdf(x, scale=1/lambda_))
plt.title('Exponential Distribution PDF')
plt.xlabel('x')
plt.ylabel('Density')

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, expon.cdf(x, scale=1/lambda_))
plt.title('Exponential Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import binom

# Parameters
n, p = 10, 0.5

# Values
x = np.arange(0, n+1)

# Plot PMF
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(x, binom.pmf(x, n, p))
plt.title('Binomial Distribution PMF')
plt.xlabel('x')
plt.ylabel('Probability')

# Plot CDF
plt.subplot(1, 2, 2)
plt.step(x, binom.cdf(x, n, p), where='mid')
plt.title('Binomial Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import uniform

# Parameters
a, b = 0, 1

# Values
x = np.linspace(a, b, 100)

# Plot PDF
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, uniform.pdf(x, loc=a, scale=b-a))
plt.title('Uniform Distribution PDF')
plt.xlabel('x')
plt.ylabel('Density')

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, uniform.cdf(x, loc=a, scale=b-a))
plt.title('Uniform Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import poisson

# Parameters
lambda_ = 3

# Values
x = np.arange(0, 15)

# Plot PMF
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(x, poisson.pmf(x, lambda_))
plt.title('Poisson Distribution PMF')
plt.xlabel('x')
plt.ylabel('Probability')

# Plot CDF
plt.subplot(1, 2, 2)
plt.step(x, poisson.cdf(x, lambda_), where='mid')
plt.title('Poisson Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Collect Historical Stock Price Data
# For this example, we'll create synthetic data for simplicity
np.random.seed(0)
dates = pd.date_range('2022-01-01', periods=100)
prices = np.cumprod(1 + np.random.normal(0.001, 0.02, len(dates)))

# Step 2: Calculate Daily Returns
prices_series = pd.Series(prices, index=dates)
returns = prices_series.pct_change().dropna()


# Plot the histogram of the returns
plt.figure(figsize=(10, 6))
plt.hist(returns, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')


# # Step 3: Estimate Parameters
mu, sigma = returns.mean(), returns.std()

# # Step 4: Plot the Parametric Density Estimate
# # Values for the x-axis
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# # Plot the normal distribution PDF
plt.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=2)
plt.title('Parametric Density Estimation for Stock Returns')
plt.xlabel('Daily Returns')
plt.ylabel('Density')
plt.legend(['Normal PDF', 'Histogram of Returns'])
plt.show()

```
