```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd

# Fetching historical data for BTC/USDT (assuming you want daily closing prices)
btc_data = yf.download('BTC-USD', start='2023-01-01', end='2023-08-01')

# Extract the 'Close' prices
X = np.arange(len(btc_data)).reshape(-1, 1)  # Time index as X
y = btc_data['Close'].values  # Closing prices as y

# Manual calculation of slope (m) and intercept (b)
mean_x = np.mean(X)
mean_y = np.mean(y)

# Calculate the slope (m) and intercept (b) using the formulas for linear regression
numerator = np.sum((X - mean_x) * (y - mean_y).reshape(-1, 1))
denominator = np.sum((X - mean_x)**2)
m_manual = numerator / denominator
b_manual = mean_y - m_manual * mean_x

# Manual prediction using calculated m and b
y_pred_manual = m_manual * X + b_manual

# Using scikit-learn to fit the model
model = LinearRegression()
model.fit(X, y)
y_pred_sklearn = model.predict(X)

# Print the results to compare
print(f"Manual calculation: m = {m_manual}, b = {b_manual}")
print(f"Scikit-learn calculation: m = {model.coef_[0]}, b = {model.intercept_}")

```
