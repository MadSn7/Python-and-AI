```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated underlying prices
underlying_prices = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

# Option strike prices
strike_call = 95
strike_put = 110

# Calculate payoffs for call and put options
payoff_call = np.maximum(underlying_prices - strike_call, 0)
payoff_put = np.maximum(strike_put - underlying_prices, 0)

# Define colors based on condition
colors_call = ['green' if price >= strike_call else 'red' for price in underlying_prices]
colors_put = ['green' if price <= strike_put else 'red' for price in underlying_prices]

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot for call options with connected lines
for i in range(len(underlying_prices) - 1):
    plt.plot(underlying_prices[i:i+2], payoff_call[i:i+2], marker='o', color=colors_call[i])

# Plot for put options with connected lines
for i in range(len(underlying_prices) - 1):
    plt.plot(underlying_prices[i:i+2], payoff_put[i:i+2], marker='o', color=colors_put[i])

# Highlighting the strike prices
plt.axvline(x=strike_call, color='blue', linestyle='--', label=f'Call Strike Price ({strike_call})')
plt.axvline(x=strike_put, color='purple', linestyle='--', label=f'Put Strike Price ({strike_put})')

plt.title('Option Payoff at Expiration with Intrinsic Values and Connected Lines')
plt.xlabel('Underlying Price')
plt.ylabel('Payoff')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated underlying prices
underlying_prices = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

# Option strike prices
strike_call = 95
strike_put = 110

# Premium costs for the options
premium_call = 5  # Example premium for the call option
premium_put = 7   # Example premium for the put option

# Calculate payoffs for call and put options
payoff_call = np.maximum(underlying_prices - strike_call, 0) - premium_call
payoff_put = np.maximum(strike_put - underlying_prices, 0) - premium_put

# Define colors based on net payoff
colors_call = ['green' if payoff >= 0 else 'red' for payoff in payoff_call]
colors_put = ['green' if payoff >= 0 else 'red' for payoff in payoff_put]

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot for call options with connected lines
for i in range(len(underlying_prices) - 1):
    plt.plot(underlying_prices[i:i+2], payoff_call[i:i+2], marker='o', linestyle='-', color=colors_call[i])

# Plot for put options with connected lines
for i in range(len(underlying_prices) - 1):
    plt.plot(underlying_prices[i:i+2], payoff_put[i:i+2], marker='o', linestyle='-', color=colors_put[i])

# Highlighting the strike prices
plt.axvline(x=strike_call, color='blue', linestyle='--', label=f'Call Strike Price ({strike_call})')
plt.axvline(x=strike_put, color='purple', linestyle='--', label=f'Put Strike Price ({strike_put})')

plt.title('Option Payoff at Expiration Including Premium Costs')
plt.xlabel('Underlying Price')
plt.ylabel('Net Payoff')
plt.grid(True)
plt.legend()
plt.show()
```
