```python
import yfinance as yf

# Set the ticker
ticker = "AAPL"

# Get the data on this ticker
data = yf.Ticker(ticker)

# Get options expirations
options_dates = data.options

# Fetch the option chain for each expiration date
for date in options_dates:
    options = data.option_chain(date)
    print(f"Options for {date}:")
    # print("Calls:")
    # print(options.calls)  # Display call options
    print("Puts:")
    print(options.puts)  # Display put options
```
