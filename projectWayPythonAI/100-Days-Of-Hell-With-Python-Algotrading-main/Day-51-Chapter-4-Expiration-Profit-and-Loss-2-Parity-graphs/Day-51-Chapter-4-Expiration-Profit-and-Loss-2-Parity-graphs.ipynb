

### 1. Long Call
**Explanation:**
A long call position becomes profitable when the underlying asset's price exceeds the strike price plus the premium paid. The slope of a long call graph is zero when the underlying price is below the strike price (since the option is not exercised) and becomes positive as the price rises above the strike price.

**Examples:**
- **Underlying Price = $80, Strike = $75, Premium = $5**: Break-even at $80. No profit or loss.
- **Underlying Price = $90, Strike = $75, Premium = $5**: Profit of $10.
- **Underlying Price = $70, Strike = $75, Premium = $5**: Loss of $5 (total premium paid).






```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
underlying_prices = np.linspace(70, 90, 100)
strike_price = 75
premium = 5

# Calculate payoffs
payoff_long_call = np.maximum(underlying_prices - strike_price, 0) - premium
colors = ['green' if payoff > 0 else 'red' for payoff in payoff_long_call]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(underlying_prices, payoff_long_call, color='black')  # Base line for zero
for i in range(len(underlying_prices)-1):
    plt.plot(underlying_prices[i:i+2], payoff_long_call[i:i+2], linestyle='-', marker='o', color=colors[i])
plt.title('Long Call Option Payoff')
plt.xlabel('Underlying Price')
plt.ylabel('Profit / Loss')
plt.axhline(0, color='grey', lw=1)  # Add zero line for reference
plt.grid(True)
plt.show()
```




### 2. Short Call
**Explanation:**
A short call position is profitable when the underlying price remains below the strike price. The position incurs increasing losses as the price rises above the strike price. The slope of the short call is zero when below the strike price and negative above it.

**Examples:**
- **Underlying Price = $80, Strike = $75, Premium = $5**: Loss of $5.
- **Underlying Price = $70, Strike = $75, Premium = $5**: Profit of $5.
- **Underlying Price = $90, Strike = $75, Premium = $5**: Loss of $10.



```python
# Parameters are the same as the long call

# Calculate payoffs
payoff_short_call = premium - np.maximum(underlying_prices - strike_price, 0)
colors = ['red' if payoff < 0 else 'green' for payoff in payoff_short_call]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(underlying_prices, payoff_short_call, color='black')  # Base line for zero
for i in range(len(underlying_prices)-1):
    plt.plot(underlying_prices[i:i+2], payoff_short_call[i:i+2], linestyle='-', marker='o', color=colors[i])
plt.title('Short Call Option Payoff')
plt.xlabel('Underlying Price')
plt.ylabel('Profit / Loss')
plt.axhline(0, color='grey', lw=1)  # Add zero line for reference
plt.grid(True)
plt.show()
```


### 3. Long Put
**Explanation:**
A long put position becomes profitable when the underlying asset's price falls below the strike price minus the premium paid. The slope of a long put graph is zero when the underlying price is above the strike price (since the option is not exercised) and becomes negative as the price falls below the strike price.





```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
underlying_prices = np.linspace(70, 90, 100)
strike_price = 85
premium = 5

# Calculate payoffs
payoff_long_put = np.maximum(strike_price - underlying_prices, 0) - premium
colors = ['green' if payoff > 0 else 'red' for payoff in payoff_long_put]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(underlying_prices, payoff_long_put, color='black')  # Base line for zero
for i in range(len(underlying_prices)-1):
    plt.plot(underlying_prices[i:i+2], payoff_long_put[i:i+2], linestyle='-', marker='o', color=colors[i])
plt.title('Long Put Option Payoff')
plt.xlabel('Underlying Price')
plt.ylabel('Profit / Loss')
plt.axhline(0, color='grey', lw=1)  # Add zero line for reference
plt.grid(True)
plt.show()
```
