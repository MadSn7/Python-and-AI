```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
S0 = 100  # Initial stock price
K = 105  # Strike price of the call option
C = 5    # Premium received for the call option

# Range of final stock prices

ST = np.linspace(50, 150, 500)

# Calculate final payoff (f_T)
f_T = ST - S0 - np.maximum(0, ST - K) + C

# Calculate adjusted stock price (S_*)
S_star = S0 - C

# Calculate maximum profit (P_max)
P_max = K - S0 + C

# Calculate maximum loss (L_max)
L_max = S0 - C

# Plot the payoff diagram
plt.figure(figsize=(10, 6))
plt.plot(ST, f_T, label='Covered Call Payoff')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(S0, color='grey', linestyle='--', linewidth=0.5, label='Initial Stock Price')
plt.axvline(K, color='red', linestyle='--', linewidth=0.5, label='Strike Price')
plt.axhline(P_max, color='green', linestyle='--', linewidth=0.5, label='Max Profit')
plt.axhline(L_max, color='orange', linestyle='--', linewidth=0.5, label='Max Loss')

plt.title('Covered Call Strategy Payoff Diagram')
plt.xlabel('Stock Price at Expiration ($S_T$)')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)
plt.show()

# Display the calculated values
print(f"Adjusted Stock Price (S_*): {S_star}")
print(f"Maximum Profit (P_max): {P_max}")
print(f"Maximum Loss (L_max): {L_max}")
```
