```python
# Python Lists
# Acts like dynamic array, can have different data types in python, in array fix size, and same values like only int, float etc
# Arrays are faster and less space occupy, cause array is stored continously and fix size, list stores pointers etc for address referencing
trade_logs = [100,"BTC",120000.00,"Buy"]
trade_logs
id(trade_logs) #memory address, can fetch for particular index as well

# Types of list
# 1. Homogeneous List: Contains elements of the same data type
homogeneous_list = [1, 2, 3, 4, 5]

# 2. Heterogeneous List: Contains elements of different data types
heterogeneous_list = [1, "two", 3.0, [4]]

# 3. Nested List: A list within another list
nested_list = [1, 2, [3, 4], 5]
# 4. Multidimensional List: A list with multiple dimensions (like a matrix)
multidimensional_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
# 5. Empty List: A list with no elements
empty_list = []

# Accessing Lists
# Accessing elements in a list is done using their index (position)
list = [1, 2, 3, 4, 5]
print(list[0])  # Output: 1
print(list[-1])  # Output: 5
# Slicing Lists
print(list[1:4])  # Output: [2, 3, 4]
print(list[:3])  # Output: [1, 2, 3]
print(list[3:])  # Output: [4, 5]
print(list[-3:-1])  # Output: [3, 4]
print(list[::2])  # Output: [1, 3, 5]
# Modifying Lists
list[0] = 10
list.append(6)
print(list)  # Output: [10, 2, 3, 4, 5, 6]

list.extend([7, 8, 9]) #always adds multiple items so make sure it's multiple
print(list)  # Output: [10, 2, 3, 4, 5, 6, 7, 8, 9]

list.insert(0, 0)
print(list)  # Output: [0, 10, 2, 3, 4, 5, 6, 7, 8, 9]

# update list
list[1] = 20
print(list)  # Output: [0, 20, 2, 3, 4, 5, 6, 7, 8, 9]
# updating with slicing
list[2:5] = [30, 40, 50]
print(list)  # Output: [0, 20, 30, 40, 50, 6, 7, 8, 9]
# Deleting elements from a list
del list[0]
print(list)  # Output: [20, 30, 40, 50, 6, 7, 8, 9]
list.remove(30)
print(list)  # Output: [20, 40, 50, 6, 7, 8, 9]
popped_element = list.pop()  # Removes and returns the last element
print(popped_element)  # Output: 9
print(list)  # Output: [20, 40, 50, 6, 7, 8]
list.clear()  # Removes all elements from the list

# Operations on Lists
# 1. Concatenation: Joining two or more lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]
concatenated_list = list1 + list2
print(concatenated_list)  # Output: [1, 2, 3, 4, 5, 6]
# 2. Arithmetic Operations: Performing operations on list elements
list3 = [7, 8, 9]
sum_list = [x + y for x, y in zip(concatenated_list, list3)]
print(sum_list)  # Output: [8, 10, 12, 4, 5, 6]
# 3. Filtering: Creating a new list with elements that meet a certain condition
filtered_list = [x for x in concatenated_list if x % 2 == 0]
print(filtered_list)  # Output: [2, 4, 6]

# List Functions
# len/max/min/sorted/count/reverse/sort/copy
print(len(concatenated_list))  # Output: 6
print(max(concatenated_list))  # Output: 9
print(min(concatenated_list))  # Output: 1
print(sorted(concatenated_list))  # Output: [1, 2, 3, 4, 5, 6, 7, 8, 9] need to be homogeneous
print(concatenated_list.count(4))  # Output: 1
print(concatenated_list.reverse())  # Output: None, whenever there is no output that means it has been applied permanently
print(concatenated_list)  # Output: [6, 5, 4, 3, 2, 1]
print(concatenated_list.sort())  # Output: None, sorts the list in place
print(concatenated_list)  # Output: [1, 2, 3, 4, 5, 6] permanent
new_list = concatenated_list.copy()  # creates a shallow copy of the list

# List Comprehensions(Easy way for creating lists)
# Syntax: [expression for item in iterable if condition]
squared_list = [x**2 for x in new_list]
print(squared_list)  # Output: [1, 4, 9, 16, 25, 36]

# Can traverse a list with itemwise or indexwise
for item in new_list:  # Itemwise traversal
    print(item) 

for index in range(len(new_list)):  # Indexwise traversal
    print(new_list[index])

# ZIP : works on any number of list of any size , will stop with min length list and ignore other gives us tuple
list_a = [1, 2, 3]
list_b = ['a', 'b', 'c']
zipped = zip(list_a, list_b)
print(list(zipped))  # Output: [(1, 'a'), (2, 'b'), (3, 'c')]
```
