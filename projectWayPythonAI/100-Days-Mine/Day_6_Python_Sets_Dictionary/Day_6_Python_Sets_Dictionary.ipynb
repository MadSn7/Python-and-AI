```python
# Sets in Python
# Sets are unordered collections of unique elements. 
# Set itself is mutable, but the elements contained in it must be immutable so can't have set of set or list can have tuples
s1 = set()
a = {"Google","Tesla","Apple"}
b = {"Microsoft","Amazon","Facebook","Google"}
# Operations
print(a.union(b))  # Union
print(a.intersection(b))  # Intersection
print(a.difference(b))  # Difference
print(a.symmetric_difference(b))  # Symmetric Difference
print(a.isdisjoint(b))  # Disjoint
# Functions
print(a.issubset(b))  # Subset
print(a.issuperset(b))  # Superset
print(set('AAPL'))
# Accessing elements
s = set('AAPL')
print(s)
print('A' in s)
print('B' in s)
# Adding and delete in set
s.add('B')
print(s)
s.remove('A')
print(s)
s.update(['C', 'D'])
print(s)
print(s.discard('B')) # no error if 'B' not found
# min max sort copy
print(min(s))
print(max(s))
print(sorted(s))

# Frozen set : immutable version of python set, read will work but not update /write
fs = frozenset(s)
print(fs)
# Set Comprehension
fs_comp = set({x for x in 'AABBC'})
print(fs_comp)

# Dictionary ; key-value pairs like map
# Mutable, unordered, changeable, keys can't be duplicated and mutable
d = {'name': 'John', 'age': 30, 'city': 'New York'}
print(d)
print(d['name'])
d['age'] = 31
print(d)
# 2 d dict
d2 = {'name': 'John', 'age': 30, 'address': {'city': 'New York', 'state': 'NY'}}
print(d2['address']['city'])
# using dict() function
d3 = dict(name='John', age=30, city='New York')
print(d3)
# Adding and removing items
d['email'] = "andna@mal.com"
print(d)
del d['city']
print(d)
# cleared = d.clear()
# editing and copying
d_copy = d.copy()
print(d_copy)

# dictionary operations
print(d.keys())
print(d.values())
print(d.items())
print(d.get('name'))
print(d.pop('age'))
print(d)
print(d.popitem()) # removes last item
print('name' in d)
print('age' in d)
getattr(d, 'name', 'Not Found') # if key not found return default value
# Dictionary comprehension
squares = {x: x*x for x in range(6)}
print(squares)
# Nested Comprehension

```
