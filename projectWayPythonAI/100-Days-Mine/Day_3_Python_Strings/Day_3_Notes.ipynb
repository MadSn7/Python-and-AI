```python
# Python Strings
# Python strings are sequences of characters enclosed in quotes. They can be single (' '), double (" "), or triple (''' ''' or """ """) quoted.
# They are unicode strings, meaning they can represent characters from any language, emoji not ASCII.
# Variable sizes and various encoding types (UTF-8, UTF-16) allow for flexible string manipulation.

#Creating Strings
single_quoted = 'Hello' 
double_quoted = "Hello" # used when inside string has this quotes '' 
triple_quoted = '''
                Hello
                Hello
            ''' #Multi Lines

# Accessing Strings
print(single_quoted)
print(single_quoted[0])  # Accessing first character
print(single_quoted[-1])  # Accessing last character
print(single_quoted[1:4])  # Slicing from index 1 to 4 last is excluded
print(single_quoted[::2])  # Step Size now every 2 will be printed
print(single_quoted[::-1])  # Reversing the string

# Editing and Deleting in Strings
# Strings are immutable, meaning they cannot be changed in place.
# To modify a string, you must create a new one.
new_string = single_quoted.replace('Hello', 'Hi')  # Replacing 'Hello' with 'Hi'
print(new_string)
del new_string
# print(new_string)  # This will raise an error since new_string is deleted
```

```python
# Operations on Strings
# Arithmetic Operations only + , * works
string1 = "Hello"
string2 = "World"
print(string1 + " " + string2)  # Concatenation
print(string1 * 3)  # Repetition

# Relational Operations  == , != , > , < , >= , <= works
print(string1 == "Hello")  # True
print(string1 != string2)  # True
print(string1 > string2)  # False (based on lexicographical order)

# Logical Operations  and , or , not  works
print(string1  and  "World")  # filled string is treated as true
print(string1 == "Hello" or string2 == "Earth")  # True
print(not string1)  # False

# Loops
for char in string1:
    print(char, end=' ')  # Prints each character in string1
print()
# Membership Operations  in , not in works
print("H" in string1)  # True
print("h" not in string1)  # True
```

```python
# String Functions
# Common String Functions
string1 = "India"
print(len(string1))  # Length of the string
print(max(string1))  # Max character in the string
print(min(string1))  # Min character in the string
print(sorted(string1))  # Sorted list of characters in the string

# Capitalize/Title/Upper/Lower/Swapcase
print(string1.capitalize())  # Capitalizes the first character
print(string1.title())  # Capitalizes the first character of each word
print(string1.swapcase())  # Swaps case of all characters
print(string1.upper())  # Converts to uppercase
print(string1.lower())  # Converts to lowercase

# Count/Find/Index
print(string1.count("i"))  # Counts occurrences of 'i'
print(string1.find("l"))  # Finds the first occurrence of 'l'
print(string1.index("n"))  # Finds the first occurrence of 'n', raises error if not found

# EndsWith/StartsWith
print(string1.endswith("a"))  # Checks if the string ends with 'a'
print(string1.startswith("I"))  # Checks if the string starts with 'I'

# Format
print("Hello, {}".format(string1))  # Formats the string, order matters

# is
print(string1.isalnum())  # Checks if all characters are alphanumeric
print(string1.isalpha())  # Checks if all characters are alphabetic
print(string1.isdigit())  # Checks if all characters are digits
print(string1.isidentifier())  # Checks if all characters are valid identifiers

# split/join
print(string1.split("i"))  # Splits the string at 'i'
print("i".join(["Ind", "a"]))  # Joins the list with 'i'
# strip/lstrip/rstrip
print(string1.strip())  # Removes whitespace from both ends
print(string1.lstrip())  # Removes whitespace from the left end
print(string1.rstrip())  # Removes whitespace from the right end
```
