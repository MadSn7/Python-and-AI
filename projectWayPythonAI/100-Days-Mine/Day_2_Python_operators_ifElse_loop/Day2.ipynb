```python
# Operators in Python

# Arithmetic Operators
a = 11
b = 5
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Int division:", a // b)
print("Modulus:", a % b)

# Comparison Operators
print("Equal to:", a == b)
print("Not equal to:", a != b)
print("Greater than:", a > b)
print("Less than:", a < b)

# Logical Operators
c = True
d = False
print("Logical AND:", c and d)
print("Logical OR:", c or d)
print("Logical NOT:", not c)

# Bitwise Operators
print("Bitwise AND:", a & b)
print("Bitwise OR:", a | b)
print("Bitwise XOR:", a ^ b)
print("Bitwise NOT:", ~a)
print("Bitwise Shift Left:", a << 1)
print("Bitwise Shift Right:", a >> 1)

# Assignment Operators
a += 5
print("After += 5:", a)
a -= 3
print("After -= 3:", a)
a *= 2
print("After *= 2:", a)
a /= 4
print("After /= 4:", a)
# no a++, a--

# Identity Operators
print("Is a identical to 5?:", a is 5)
print("Is a not identical to 5?:", a is not 5)

# Membership Operators
my_list = [1, 2, 3, 4, 5]
print("Is 3 in the list?:", 3 in my_list)
print("Is 6 not in the list?:", 6 not in my_list)

# Power
print("Power:", a ** b) 

# Floor Division
print("Floor Division:", a // b)
```

```python
#if-else statement in python

entry = 10
exit = 20
sl = 8
if entry > exit:
    print("Entry is greater than Exit")
elif entry < exit:
    print("Entry is less than Exit")
else:
    print("Entry is equal to Exit")
```

```python
# Python Modules
# Modules in Python are files containing Python code. 
# They can define functions, classes, and variables that can be reused in other Python programs.
# Some modules are built into Python, while others can be created by users.
# Examples of built-in modules include `math`, `sys`, and `os`.
import math
print(math.factorial(6))  # Returns the factorial of 6

import keyword
print(keyword.kwlist)  # Prints the list of keywords in Python

import random
random_number = random.randint(1, 100)  # Generates a random number between 1 and 100
print(random_number)


import datetime
datetime_now = datetime.datetime.now()  # Gets the current date and time
print(datetime_now)

initial_price = 100
volatility = 0.05 # assuming volatility can be -5% to +5%

volatility = random.uniform(-volatility, volatility)  # Generates a random volatility factor
print("Today's Date:", datetime_now)
print("Today's Volatility:", volatility)
print("Today's Price:", initial_price + (initial_price * volatility))

help('modules') #help for all available modules
help(random) #help for particular module
```

```python
# Loops in Python
# Python has two primitive loop commands:
# 1. while loops
# 2. for loops

# While Loop
count = 0
while count < 5:
    print("Count is:", count)
    count += 1

# For Loop
for i in range(5):
    print("Iteration:", i)

# While Else 
count = 0
while count < 5:
    print("Count is:", count)
    count += 1
else:
    print("Count has reached 5.")

# For loop over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print("Fruit:", fruit)
```
